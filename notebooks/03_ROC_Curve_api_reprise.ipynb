```python
from matplotlib import pyplot as plt
```

```python
from sklearn.datasets import load_wine
from sklearn.svm import SVC
from sklearn.metrics import RocCurveDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

# 1. Prepare Data

```python
X, y = load_wine(return_X_y=True)
```

```python
y = y == 2
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

# 2. Fit The SVC Model

```python
svc = SVC(random_state=42)
```

```python
svc.fit(X=X_train, y=y_train)
```

# 3. Fit the RandomForestClassifier Model

```python
rfc = RandomForestClassifier(n_estimators=10, random_state=42)
rfc.fit(X=X_train, y=y_train)
```

# 4. Draw the ROC Curve

```python
fig_rocs = plt.figure()
```

```python
ax_rocs = fig_rocs.add_subplot(1, 1, 1)
```

```python
RocCurveDisplay.from_estimator(svc, X_test, y_test, ax=ax_rocs)
```

```python
RocCurveDisplay.from_estimator(rfc, X_test, y_test, ax=ax_rocs)
```
